## Sraping products from amazon website

### import libraries

```python
from selenium import webdriver
import pandas as pd
import time
from selenium.common.exceptions import NoSuchElementException
```

### create object  and get the page of website

```python
obj=webdriver.Chrome()
```

```python
obj.get("https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301")
```

### get elements from the page

```python
elements=obj.find_elements("xpath","//div[@class='a-section a-spacing-base']")
```

```python
for i in elements:
    print(i.text)
```

###  lists to store the data

```python
product_names=[]
prices=[]
ratings=[]
product_images=[]
product_colors=[]
product_deliver=[]
```

#### get the name,price,image for each element and put them in dataframe

```python
# time.sleep(5)
for element in elements:

    name=element.find_element("xpath", "../div//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']").text
    try:
        price = element.find_element("xpath", ".//span[@class='a-price-whole']").text
    except NoSuchElementException:
        price = "see price options"
    rate=element.find_element("xpath", ".//div[@class='a-row a-size-small']//a//span[@class='a-size-base s-underline-text']").text
    product_names.append(name)
    prices.append(price)
    ratings.append(rate)
```

```python
df = pd.DataFrame(zip(product_names,prices,ratings), columns=['Name','price','rate'])
df
```

### get image for each element

```python
images=obj.find_elements("xpath","//div[@class='a-section aok-relative s-image-square-aspect']//img[@class='s-image']")
```

```python
for i in images:
    print(i.get_attribute("src"))
    
```

```python
for i in images:
    product_images.append(i.get_attribute("src"))
```

```python
df['product_image']=product_images
df.head()
```

#### save the data in csv file

```python
df.to_csv("amozon_products.csv")
```
